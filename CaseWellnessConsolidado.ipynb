```python

# --- Instalar e importar bibliotecas necessárias ---
from google.colab import drive
from IPython.display import display
import pandas as pd
import os
import plotly.express as px

# --- 1. Montar o Google Drive ---
drive.mount('/content/drive')

# --- 2. Definir o caminho da pasta no Drive ---
# Substitua "Colab Notebooks" pelo nome exato da pasta onde está o arquivo
pasta = '/content/drive/MyDrive/Colab Notebooks'
arquivo01 = 'resultado_consolidado.cvs.csv'
caminhoCompleto01 = os.path.join(pasta, arquivo01)

aqurvivo02= 'raw.csv'
caminhoCompleto02 = os.path.join(pasta, aqurvivo02)
```

### Lendo os arquivos

```python
### raw data, arquivo tratado na v1.

df_raw = pd.read_csv(caminhoCompleto02)
print("✅ Dados carregados com sucesso!")
print(display(df_raw.head(10)))
```

```python
### métricas consolidadas, arquivo tratado na v1.
df = pd.read_csv(caminhoCompleto01)
print("✅ Dados carregados com sucesso!")
print(display(df.head(10)))
```

### Gráfico: Média de Produtividade por Regime de Trabalho

```python
# --- 4. Cálculo da média de produtividade ---
mediaProdutividade = (
    df[df['occupation'] == 'Employed']
    .groupby(['work_mode'])['productivity_0_100']
    .median()
    .reset_index()
)

##print("\n📊 Média de produtividade por tipo de trabalho:")
##print(mediaProdutividade)

# Gráfico interativo com Plotly ---
grafico01 = px.bar(
    mediaProdutividade,
    x='work_mode',
    y='productivity_0_100',
    title='Média de Produtividade por Modo de Trabalho',
    text_auto=True,
    color='work_mode'
)

grafico01.show()


mediaProdutividadeGenero = (
    df[df['occupation']=='Employed']
    .groupby(['work_mode','gender'])['productivity_0_100']
    .median()
    .reset_index()
)

grafico002 = px.line(
    mediaProdutividadeGenero,
    x='work_mode',
    y='productivity_0_100',
    color='gender',
    title='Média de Produtividade por Gênero',
    markers=True,   # adiciona pontos nas linhas
    text='productivity_0_100'
)
grafico002.show()
```
