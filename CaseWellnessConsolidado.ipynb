###Case: Screen Time vs Mental Wellness Survey
Obetivo:

*  Analisar de que forma as atividades diárias — influenciam a produtividade de profissionais em diferentes regimes de trabalho: híbrido, presencial e remoto.

1.   Média de produtividade por regime de trabalho
2.   Analisar a relação entre o tempo de prática de exercícios físicos e o regime de trabalho adotado.
3.  Entender a relação entre leisure_screen_hours e o stress_level entre os diferentes regimes dee trabalho



###1: Importar bibliotecas e montar acesso google drive

```python

# --- Instalar e importar bibliotecas necessárias ---
from google.colab import drive
from IPython.display import display
import pandas as pd
import os
import plotly.express as px

# --- 1. Montar o Google Drive ---
drive.mount('/content/drive', force_remount=True)

# --- 2. Definir o caminho da pasta no Drive ---
# Substitua "Colab Notebooks" pelo nome exato da pasta onde está o arquivo
pasta = '/content/drive/MyDrive/Colab Notebooks'
arquivo01 = 'resultado_consolidado.csv'
caminhoCompleto01 = os.path.join(pasta, arquivo01)

aqurvivo02= 'raw.csv'
caminhoCompleto02 = os.path.join(pasta, aqurvivo02)
```

### Lendo os arquivos

```python
### raw data, arquivo tratado na v1.

df_raw = pd.read_csv(caminhoCompleto02)
print("✅ Dados carregados com sucesso!")
print(display(df_raw.head(10)))
```

```python
### métricas consolidadas, arquivo tratado na v1.
df = pd.read_csv(caminhoCompleto01)
print("✅ Dados carregados com sucesso!")
print(display(df.head(10)))
```

### Gráfico: Média de Produtividade por Regime de Trabalho

```python
# --- 4. Cálculo da média de produtividade ---
mediaProdutividade = (
    df_raw[df_raw['occupation'] == 'Employed']
    .groupby(['work_mode'])['productivity_0_100']
    .median()
    .reset_index()
)

##print("\n📊 Média de produtividade por tipo de trabalho:")
##print(mediaProdutividade)

grafico01 = px.bar(
    mediaProdutividade,
    x='work_mode',
    y='productivity_0_100',
    color='work_mode',
    title='Rank com a Média de Produtividade por Regime de Trabalho',
    text_auto=True,
    range_y=[0, 100]
)
grafico01.show()

mediaProdutividadeGenero = (
    df[df['occupation']=='Employed']
    .groupby(['work_mode','gender'])['productivity_0_100']
    .median()
    .reset_index()
)

grafico002 = px.line(
    mediaProdutividadeGenero,
    x='work_mode',
    y='productivity_0_100',
    color='gender',
    title='Média com o Rank de Produtividade por Gênero',
    markers=True,   # adiciona pontos nas linhas
    text='productivity_0_100',
    range_y=[0, 100]
)
grafico002.show()
```

```python
# --- 5. Analisar a relação entre o tempo de prática de exercícios físicos e o rank de produtividade
mediaExercicio = (
    df_raw[df_raw['occupation'] == 'Employed']
    .groupby(['work_mode'])['exercise_minutes_per_week']
    .median()
    .reset_index()
)

##print("\n📊 Média de produtividade por tipo de trabalho:")
##print(mediaProdutividade)

# Gráfico interativo com Plotly ---
grafico03 = px.bar(
    mediaExercicio,
    x='work_mode',
    y='exercise_minutes_per_week',
    title='Média de Exercício Físico Por Regime de Trabalho',
    text_auto=True,
    color='work_mode'
)

grafico03.show()

mediaProdutividade = (
    df_raw[df_raw['occupation'] == 'Employed']
    .groupby(['work_mode'])['productivity_0_100']
    .median()
    .reset_index()
)

grafico04 = px.bar(
    mediaProdutividade,
    x='work_mode',
    y='productivity_0_100',
    title='Rank de Produtividade Por Regime de Trabalho',
    text_auto=True,
    color='work_mode',
    range_y=[0, 100]
)

grafico04.show()

```

```python
# --- 6. Correlação da prática de exercício fisico com a saude mental por regime de trabalho adotado

df_saudeMental = df_raw[df_raw['occupation'] == 'Employed'][[
    'user_id', 'work_mode', 'exercise_minutes_per_week', 'mental_wellness_index_0_100'
]]

graficoSaudeMental = px.scatter(
    df_saudeMental,
    x='exercise_minutes_per_week',
    y='mental_wellness_index_0_100',
    ##color='work_mode',            # Mantém cores diferenciadas (opcional)
    hover_data=['user_id'],
    facet_col='work_mode',        # Cria um painel separado para cada work_mode
    trendline='ols',              # adiciona linha de tendência linear
    title='Exercício x Bem-estar Mental por Usuário',
    range_y=[0,100]
)

graficoSaudeMental.show()
```

```python
# --- 6. Correlação de nível de stress  por regime de trabalho adotado

df_saudeMentalStress = df_raw[df_raw['occupation'] == 'Employed'][[
    'user_id', 'work_mode', 'stress_level_0_10', 'screen_time_hours', 'mental_wellness_index_0_100'
]]

graficoSaudemetalStress = px.scatter(
df_saudeMentalStress,
x='screen_time_hours',
y='stress_level_0_10',
hover_data=['user_id'],
facet_col='work_mode',
trendline='ols',
title='Nivel de Stress Por Regime de Trabalho',
range_y=[0,12]
)

graficoSaudemetalStress.show()

mediaScreenTime = (
    df_raw[df_raw['occupation'] == 'Employed']
    .groupby(['work_mode'])[['screen_time_hours','work_screen_hours','leisure_screen_hours']]
    .median()
    .reset_index()
)

print(mediaScreenTime)
```
